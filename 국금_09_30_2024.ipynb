```python
df = pd.read_csv("world_gdp_current.csv")
```

```python
countrycode = df.iloc[:, 0:2]
countrycode.head()
```

```python
df.head()
```

```python
df.drop(columns = ["Indicator Name", "Indicator Code", "Country Code"], inplace = True)
```

```python
df.head()
```

```python
df_melt = df.melt(id_vars = ["Country Name"], var_name = "Year", value_name = "GDP")
df_melt.head()
```

```python
df_melt
```

```python
df_melt.rename(columns = {"Country Name" : "Country"}, inplace = True)
df_melt
```

```python
df_melt["GDP_million"] = df_melt["GDP"].div(1000000)
df_melt.head()
```

```python
df_merged = df_melt.merge(countrycode, left_on = "Country", right_on = "Country Name")
df_merged
```

```python
df_merged["Country"].unique()
```

```python
options = ["Iceland", "Italy", "United States", "Japan", "Korea, Rep.", "Zimbabwe"]
df_option = df_merged[df_merged["Country"].isin(options)]
```

```python
df_merged["Country"].isin(options) # 이렇게하면  boolean 값
```

```python
df_options = df_merged[df_merged["Country"].isin(options)] # 해당하는 국가만 나옴
```

```python
df_options
```

```python
# time series 시계열
# 6개 국가를 time series로 yearly 봤습니다. (frequency 얘기) 1980년부터 2022년까지
df_options["Year"] = pd.to_datetime(df_options["Year"]) # Year 변수를 날짜 형식 to_datetime 으로 형변환
```

```python
df_options
```

```python

df_options_1980 = df_options.loc[df_options["Year"] >= "1980"] # 1980 이상, df_options["Year"] 만 하면 boolean값 나옴
```

```python
df_options_1980
```

```python
#import seaborn as sns 앞에서 이미 선언함
```

```python
plt.figure(figsize = (15, 5)) # 사이즈 조정
sns.lineplot(x = "Year", y = "GDP_million", data = df_options_1980) # shade로 나옴
```

```python
plt.figure(figsize = (15, 5))
sns.lineplot(x = "Year", y = "GDP_million", data = df_options_1980, hue = "Country Name", marker = "o") # hue 국가 이름에 따라 따로 선으로, marker o로 점찍음
plt.xlabel("Year")
plt.ylabel("GDP (Millions)")
# gdp 백만으로 단위를 downscale시켰음에도 짐바브웨와 us의 차이가 큼 => 성장률을 보자 percent growth rate
```
