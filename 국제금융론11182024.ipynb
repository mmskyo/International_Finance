````python
url="https://gml.noaa.gov/webdata/ccgg/trends/co2/co2_mm_mlo.csv"

````

````python
df=pd.read_csv(url, comment="#")
````

````python
df.head()
````

````python
df["date"]=pd.to_datetime(dict(year=df["year"], month=df["month"], day=1))
````

````python
df.head()
````

```
  dict(year=df["year"], month=df["month"], day=1)

```
- 연도(year), 월(month), 일(day) 값을 하나의 사전 형태로 제공
- 이 사전은 pandas 가 날짜 정보를 조합하는 데 사용
- day=1은 특정날짜가 명시되지 않을대 모든 날짜를 해당 월의 첫번째 날로 설정하는 예시   

````python
df=df[["date", "average"]]
````

````python
df.head()
````

````python
df["date"]=pd.to_datetime(df["date"])
df.set_index("date", inplace=True)
````

````python
df.head()
````

````python
df.describe()
````

````python
plt.plot(df["average"], color="brown")
````

````python
adfuller(df)
````

- KPSS test

- 귀무가설: stationary

````python
from statsmodels.tsa.stattools import kpss
````

````python
kpss(df)
````

- ADF test p-value =1.0 >0.05
   - 귀무가설을 accept

- KPSS test p-value=0.01 < 0.05
  - 귀무가설 reject

=> nonstationary

````python
from statsmodels.tsa.seasonal import seasonal_decompose
````

````python
result=seasonal_decompose(df, model="additive")
result.plot()
plt.show()
````

````python
from sktime.transformations.series.difference import Differencer
````

````python
transformer=Differencer(lags=1)
df["average"]=transformer.fit_transform(df["average"])
````

````python
df.head()
````

````python
result_diff=seasonal_decompose(df, model="additive")
result_diff.plot()
plt.show()
````

````python
from pandas.plotting import autocorrelation_plot
````

````python
df.shape
````

````python
autocorrelation_plot(df[:48])
````

- critical range $(-0.3,0.3)$

- 48개의 time unit, 12th, 24th lag 에서 통계적으로 유의미한 양(+)의 상관관계

- 6th~9th lag: 통계적으로 유의미한 음(-)의 약한 상관관계 (weakly negatie correlation)
